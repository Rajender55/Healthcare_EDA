**Using fillna method to process the missing data**

```python
df['Population'] = df['Population'].fillna(df['Male'] + df['Female'])
df['Literate'] = df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'])
df['Households']=df['Households'].fillna(df['Households_Urban']+df['Households_Rural'])
df['Male'] = df['Male'].fillna(df['Population'] - df['Female'])
df['Female'] = df['Female'].fillna(df['Population'] - df['Male'])
df['Literate_Male'] = df['Literate_Male'].fillna(df['Literate'] - df['Literate_Female'])
df['Literate_Female'] = df['Literate_Female'].fillna(df['Literate'] - df['Literate_Male'])
df['Households_Rural']=df['Households_Rural'].fillna(df['Households']-df['Households_Urban'])
df['Households_Urban']=df['Households_Urban'].fillna(df['Households']-df['Households_Rural'])

df
```

```python
df.isna().sum()
```

```python
df['Young_and_Adult'] = df['Young_and_Adult'].fillna(df['Population'] -(df['Middle_Aged']+df['Senior_Citizen']+df['Age_Not_Stated']))
df['Middle_Aged'] = df['Middle_Aged'].fillna(df['Population'] -(df['Young_and_Adult']+df['Senior_Citizen']+df['Age_Not_Stated']))
df['Senior_Citizen'] = df['Senior_Citizen'].fillna(df['Population'] -(df['Young_and_Adult']+df['Middle_Aged']+df['Age_Not_Stated']))
df['Age_Not_Stated'] = df['Age_Not_Stated'].fillna(df['Population'] -(df['Young_and_Adult']+df['Middle_Aged']+df['Senior_Citizen']))
df
```

# Problem Statement 17: (Hospitals required to meet the standards)

**Find the average number of beds in a government hospital. Divide it by the gap in the number of beds to reach
the standards in the region, and round the number to the nearest integer to find the value.**

```python
df_hos_13=pd.read_csv(r'C:\Users\Futurense\Downloads\DS - Python + SQL + ETL - Healthcare Census\DS - Python + SQL + ETL - Healthcare Census\Clean_data\government_hospital.csv')
df_hos_13.head()
```

```python
df_17=private_gov[['State/UT','Rural_Government_Hospitals','Rural_Government_Beds','Urban_Government_Hospitals',
                   'Urban_Government_Beds','Total_number_of_government_hospital_beds','Hospitals','HospitalBeds','Population',
                  'Sum_of_private_and_gov_hospital_beds','Hospitalbeds_per_1000_people']]
df_17.head()
```

```python
df_17['Total_gov_private_hospitals']=df_17['Hospitals']+df_17['Urban_Government_Hospitals']+df_17['Rural_Government_Hospitals']
```

```python
#Required number of hospital beds as per WHO standards
df_17['Req_num_of_beds_as_per_WHO']=round((df_17['Population']*3)/1000)
```

```python
#Calculate the avg number of beds per hospital
df_17['Avg_num_of_hospital_beds_per_hospital']=round(df_17['Sum_of_private_and_gov_hospital_beds']/df_17['Total_gov_private_hospitals'])
```

```python
df_17.head(3)
```

```python
df_17['Req_no_of_Hosp_WHO']=round(df_17['Req_num_of_beds_as_per_WHO']/df_17['Avg_num_of_hospital_beds_per_hospital'])
```

```python
df_18=df_17[df_17['Total_gov_private_hospitals']<df_17['Req_no_of_Hosp_WHO']]
```

```python
df_18[['State/UT','Total_gov_private_hospitals','Req_no_of_Hosp_WHO']].head(5)
```

```python
df_18['New_Hos_to_add_asper_WHO']=df_18['Req_no_of_Hosp_WHO']-df_18['Total_gov_private_hospitals']
```

```python
#The decision need to be taken for each state
df_17_final=df_18[['State/UT','Total_gov_private_hospitals','Req_no_of_Hosp_WHO','New_Hos_to_add_asper_WHO']]
```

```python
df_17_final.sort_values(by='New_Hos_to_add_asper_WHO', ascending=False, inplace=True)
df_17_final
```

```python
y=df_17_final['State/UT']
x=df_17_final['New_Hos_to_add_asper_WHO']

plt.barh(y,x)
plt.xlabel('New Hospitals to be added as per WHO')
plt.ylabel('State/UT')
plt.xticks(rotation=90)

plt.show()
```
